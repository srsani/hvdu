## Get data:

```python
p = pathlib.Path("../dataset/raw/nist_data")
p.mkdir(parents=True, exist_ok=True)
```

```python
for folder in ['hsf_0', 'hsf_1', 'hsf_2', 'hsf_3']:
    for image_path in glob.glob(f'../dataset/raw/nist/hsf_page/{folder}/*.png'):

        image_name = image_path[image_path.rfind('/')+1:]
        
        im = Image.open(image_path)
#         width, height = im.size
        im1 = im.crop((100, 300, 2560, 650))
        
        im1.save(f'../dataset/raw/nist/nist_data/{image_name}') 
```
