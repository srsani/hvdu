```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('test.csv')
print(df1.shape)
df1
```

```python
result_dict = {}

fla = df1[df1.total_accuracy == 100].shape[0]/df1.shape[0]
result_dict['fla'] = np.round(fla, 4)*100

result_dict['CBA TOTAL FULL DATASET'] = np.round(df1['total_accuracy'].mean(), 4)
result_dict['CBA DATE FULL DATASET'] = np.round(df1['date_accuracy'].mean(), 4)
result_dict['CBA CITY FULL DATASET'] = np.round(df1['city_accuracy'].mean(), 4)
result_dict['CBA STATE FULL DATASET'] = np.round(df1['state_accuracy'].mean(), 4)
result_dict['CBA ZIP FULL DATASET'] = np.round(df1['zip_accuracy'].mean(), 4)

result_dict
```

```python
from pathlib import Path

def ensure_folder_exists(folder_path):
    path = Path(folder_path)
    if not path.exists():
        path.mkdir(parents=True, exist_ok=True)
        print(f"Folder '{folder_path}' created!")
    else:
        print(f"Folder '{folder_path}' already exists.")
```

```python
ensure_folder_exists('../dataset/result/nist_form/')
```

```python
import json
with open('../dataset/result/nist_form/data.json', 'w') as f:
    json.dump(result_dict, f)
```
